```python
import numpy as np
from tqdm import tqdm
import pandas as pd
from sklearn import svm,preprocessing
from sklearn.linear_model import LogisticRegression,LinearRegression,BayesianRidge,Ridge,Lasso,MultiTaskLasso,ElasticNet,LassoLars
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier,RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt


TEST_DIR=""
df=pd.read_csv(r'C:\Users\My PC\Desktop\Machine Learning\Hackerearth\Employeechurn\train.csv')
#print(df.head())

def check_broken_values(df):
	#print(df.columns)
	columns=[]
	empty=[]
	columns=df.columns
	#print(columns)
	for feature in columns:
		for rows in df[feature]:
			if pd.isna(rows)==True:
				#print(feature,"NULL_VAL_FOUND")
				empty.append(feature)
				break
				
	return empty

def fix_broken_values(df):

	empty_columns=check_broken_values(df)
	#print(empty_columns)
	for col in empty_columns:
		#print(col,np.mean(df[col]))
		df[col].fillna(np.mean(df[col]),inplace=True)
	
	#df.to_csv('final.csv',index=False)
	return df


def handle_non_numeric_data(df):
	columns=df.columns.values
	print(columns)
	for column in columns:
		text_digit_vals={}
		def convert_to_int(val):
			return text_digit_vals[val]
		if df[column].dtype!=np.int64 and df[column].dtype!=np.float64:
			column_contents=df[column].values.tolist()
			unique_elements=set(column_contents)

			x=0
			for unique in unique_elements:
				if unique not in text_digit_vals:
					text_digit_vals[unique]=x
					x+=1
			df[column]=list(map(convert_to_int,df[column]))
	#df.to_csv('finale.csv',index=False)
	return df


#df1=fix_broken_values(df)
#print(check_broken_values(df1))
#df1=handle_non_numeric_data(df1)
#print(df1.head())

#df=pd.read_csv('finale.csv')
#print(df.head())
#df=pd.read_csv(TRAIN_DIR)
df=fix_broken_values(df)
df=handle_non_numeric_data(df)
df.drop(['Employee_ID'],1,inplace=True)
x=np.array(df.drop(['Attrition_rate'],1).astype(float))
#x=preprocessing.scale(x)
print(x)
y=np.array(df.Attrition_rate)
clf=GradientBoostingRegressor()
# train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=.3,random_state=1993)














# plot feature importance

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
clf.fit(x_train,y_train)
'''print(clf.feature_importances_)
plot_importance(clf)'''
#print(x)
#plt.show()
accuracy=clf.score(x_test,y_test)
print(accuracy)
predict_me=np.array([1,42.0,4,1,4,5,1,4.0,4,33,1,1,7.0,4,3.0,4,0.7516,1.8688,2.0,4,5,3])
predict_me=predict_me.reshape(-1,len(predict_me))
#print(clf.predict(predict_me))
df4=pd.read_csv(r'C:\Users\My PC\Desktop\Machine Learning\Hackerearth\Employeechurn\test.csv')
df3=pd.read_csv(r'C:\Users\My PC\Desktop\Machine Learning\Hackerearth\Employeechurn\test.csv')
#print(df2.head())
with open(r'C:\Users\My PC\Desktop\Machine Learning\Hackerearth\Employeechurn\sub.csv','w') as f:
	f.write('Employee_ID,Attrition_rate\n')
df2=fix_broken_values(df3)
df2=handle_non_numeric_data(df2)


for count,rows in enumerate(df2.Age):
	predict_me=np.array([df2.Gender[count],df2.Age[count],df2.Education_Level[count],df2.Relationship_Status[count],df2.Hometown[count],df2.Unit[count],df2.Decision_skill_possess[count],df2.Time_of_service[count],df2.Time_since_promotion[count],df2.growth_rate[count],df2.Travel_Rate[count],df2.Post_Level[count],df2.Pay_Scale[count],df2.Compensation_and_Benefits[count],df2.Work_Life_balance[count],df2.VAR1[count],df2.VAR2[count],df2.VAR3[count],df2.VAR4[count],df2.VAR5[count],df2.VAR6[count],df2.VAR7[count]])
	predict_me=predict_me.reshape(-1,len(predict_me))
		#print(clf.predict(predict_me))
	with open(r'C:\Users\My PC\Desktop\Machine Learning\Hackerearth\Employeechurn\sub.csv','a') as f:
		f.write("{},{}\n".format(df4.Employee_ID[count],clf.predict(predict_me)[0]))
```
